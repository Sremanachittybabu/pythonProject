```python
import pandas
from datetime import datetime
from pytz import utc
import matplotlib.pyplot as plt
data = pandas.read_csv("reviews.csv" , parse_dates = ['Timestamp'])
```

```python
dayavg = data.groupby('Timestamp')
dayavg.head()
```

```python
data['Date'] = data['Timestamp'].dt.date
data.head()
```

```python
data['Day'] = data['Timestamp'].dt.date
dayavg = data.groupby(['Day']).mean('Day')
plt.plot(dayavg.index , dayavg['Rating'])
```

```python
data['weekly'] = data['Timestamp'].dt.strftime('%Y-%U')
avg = data.groupby(['weekly']).mean('weekly')
plt.plot(avg.index,avg['Rating'])
```
