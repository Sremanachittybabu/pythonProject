```python
import pandas
from datetime import datetime
from pytz import utc
import matplotlib.pyplot as plt
data = pandas.read_csv("reviews.csv" , parse_dates = ['Timestamp'])
```

```python
data['Day'] = data['Timestamp'].dt.date
dayavg = data.groupby(['Day']).mean('Day')
plt.plot(dayavg.index , dayavg['Rating'])
```

```python
data['Month'] = data['Timestamp'].dt.strftime('%Y-%m')
avg = data.groupby(['Month']).mean('Month')
plt.plot(avg.index,avg['Rating'])
```

```python
#Avg rating by month by course
data['Month'] = data['Timestamp'].dt.strftime('%Y-%m')
avg1 = data.groupby(['Month' ,'Course Name'])['Rating'].count().unstack()
avg1.plot()
```
